```python
import pandas as pd
from matplotlib import pyplot as plt
import os
from os import path as osp
from PIL import Image
```

```python
some_dots = pd.read_csv("./heatmap_results/bag.csv", header=None)
```

```python
some_dots.values[0]
```

```python
images = sorted([x for x in os.listdir("/data/vot2016/bag") if x.endswith("jpg")])
```

```python
ind = 0
for i in range(len(images)):
    plt.imshow(Image.open(osp.join("/data/vot2016/bag/", images[i])))
    index, dotx, doty = some_dots.values[ind]
    if index != i: continue
    plt.plot([dotx], [doty], 'ro')
    ind += 1
    plt.title(images[i])
```
